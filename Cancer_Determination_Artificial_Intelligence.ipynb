```python
import tensorflow as tf
#library to create a basic neural network. Input layer -> hidden layer -> Output layer

import pandas as pd
#Pandas is a Python library that facilitates data manipulation and analysis through its powerful data structures and functions
#pandas now has the shortform pd


from sklearn.model_selection import train_test_split
#library that splits dataset into a testing set and a dataset, done to test model on data it hasnt seen before

from google.colab import drive
#imports drive module which enables us to mount google drive files into the project enviroment
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/cancer.csv')
#this assigns the data file in my google drive cancer.csv to the variable dataset
```

```python
x = dataset.drop(columns = ["diagnosis(1=m, 0=b)"])
#sets the x axis in the data to be every other column except "diagnosis(1=m, 0=b)"

y = dataset["diagnosis(1=m, 0=b)"]
#sets the y axis as "diagnosis(1=m, 0=b)"
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
#this splits the data, making 30% of data into testing set
```

We will now create the neural network

```python
model = tf.keras.models.Sequential()
#this creates the neural network
```

We are now adding layers to the neural network

```python
#input layer
model.add(tf.keras.layers.Dense(128, input_dim = x_train.shape[1]))
#This adds a layer to the neural network with function Dense(), we added 256, unit which is power to 2
#the input part shows that the input will be the size of the x rows specificed earlier, which is quite big
#the activation makes a sigmoid! this puts all the values in the neural network in between 0 and.1

#this is where the hidden layer starts
#hidden layer 1
model.add(tf.keras.layers.Activation('sigmoid'))
model.add(tf.keras.layers.Dropout(0.4))
#hidden layer 2
model.add(tf.keras.layers.Dense(128))
model.add(tf.keras.layers.Activation('sigmoid'))
model.add(tf.keras.layers.Dropout(0.4))
#hidden layer 3
model.add(tf.keras.layers.Dense(64))
model.add(tf.keras.layers.Activation('sigmoid'))
model.add(tf.keras.layers.Dropout(0.4))
#hidden layer 4
model.add(tf.keras.layers.Dense(32))
model.add(tf.keras.layers.Activation('sigmoid'))
model.add(tf.keras.layers.Dropout(0.4))
#hidden layer 5
model.add(tf.keras.layers.Dense(16))
model.add(tf.keras.layers.Activation('sigmoid'))
model.add(tf.keras.layers.Dropout(0.4))

#sigmoid is a function that allows values in range 0 to 1, this will reduce the complexity of the algorithm
#.Droupout(0.4) is a regularization technique to prevent overfitting, this will randomly set 40% of inputs to 0 at each update during training time
#The decrease in number of nuerons throught the .Dense(#) starting from 128 and ending at 16, is called a bottle neck structure, this allows for the algorthm to focus on more fine details in the dataset staring with high level features and ending with low level features



#output layer
model.add(tf.keras.layers.Dense(1))
#value 1 is used as the output should be boolean to show if cell is cancerous or not
```

```python
model.summary()
#this allows us to quickly asses the structure of our AI model
```

```python
#now to compile model
model.compile(optimizer = 'adam', loss = tf.keras.losses.mean_squared_error, metrics = ['accuracy'])

#Optimizer adam is how the algorithm is optimized
#The loss function uses metric called binary_crossentropy, THIS IS GOOD FOR CATEGORIZING DESCRETE THINGS, such as Cancerous or non-cancerous
#metrics is what we want to observe in our model and specifically the models accuracy
```

```python
#now fit data into the input layer

model.fit(x_train , y_train, epochs = 250)
#x_train is ran over a y_train column
#epochs is number of times the algorithm is iterating through the same data
```

```python
model.evaluate(x_test, y_test, batch_size = 64)
#this is used to evalulate the performance of our AI model
```
